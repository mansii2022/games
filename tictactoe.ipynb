```python
!apt-get install -y xvfb python3-tk  # Install xvfb
!pip install pyvirtualdisplay # Install the pyvirtualdisplay module

from tkinter import *
import tkinter.messagebox as msg
from pyvirtualdisplay import Display

# Create a virtual display
display = Display(visible=0, size=(800, 600))
display.start()

# Initialize the main window
root = Tk()
root.title('TIC-TAC-TOE---DataFlair')

# Global variables
click = True
count = 0
board = [' '] * 10

# Function to check for a win
def check_win():
    win_conditions = [(1, 2, 3), (4, 5, 6), (7, 8, 9), (1, 4, 7), (2, 5, 8), (3, 6, 9), (1, 5, 9), (3, 5, 7)]
    for a, b, c in win_conditions:
        if board[a] == board[b] == board[c] and board[a] != ' ':
            return True
    return False

# Function to handle button clicks
def checker(button_number):
    global click, count
    if board[button_number] == ' ':
        if click:
            board[button_number] = 'X'
            buttons[button_number].config(text='X', state=DISABLED)
        else:
            board[button_number] = 'O'
            buttons[button_number].config(text='O', state=DISABLED)
        click = not click
        count += 1

        if check_win():
            msg.showinfo("Tic-Tac-Toe", f"Player {'1' if not click else '2'} wins!")
            reset()
        elif count == 9:
            msg.showinfo("Tic-Tac-Toe", "It is a Tie!")
            reset()

# Function to reset the game
def reset():
    global board, count, click
    board = [' '] * 10
    count = 0
    click = True
    for button in buttons.values():
        button.config(text=' ', state=NORMAL)

# Create the UI
Label(root, text="Player 1: X", font="times 15").grid(row=0, column=1)
Label(root, text="Player 2: O", font="times 15").grid(row=0, column=2)

buttons = {}
for i in range(1, 10):
    buttons[i] = Button(root, width=15, height=7, font=('Times 16 bold'), command=lambda i=i: checker(i))
    buttons[i].grid(row=(i-1)//3 + 1, column=(i-1)%3 + 1)

root.mainloop()
```
